```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_ax = np.linspace(-3,6, 9)
x = np.array([0, 1, 2, 1, 1, 0, 0, 0, 0])
h = np.array([0, 0, 1, -1, 0, 0, 1, 1, 0])
y = signal.convolve(x,h)
```

```python
plt.stem(x_ax, x, label="$x[k]$")
plt.xlabel('$k$')
plt.legend(loc='best')
plt.ylim(-2,5)
plt.show()
```

```python
for i in range(-2,8):
    plt.stem(i-x_ax, h, label='h['+str(i)+'-k]')
    plt.xlabel('$k$')
    plt.legend(loc='best')
    plt.ylim(-2,5)
    plt.xlim(-10,11)
    plt.show()
```

```python
x_ax_y = np.linspace(-5,11,17)
plt.stem(x_ax_y,y, label='y[n]')
plt.title('Discrete Convolution')
plt.xlabel('$n$')
plt.show()
```
